# implementing PCA

```python
# principle component analysis
from sklearn.decomposition import PCA

pca = PCA(0.95)
x_pca= pca.fit_transform(x)


print("PCA shape of x is:",x_pca.shape, "and original shape is :", x.shape)
print("% of importance of selected features is :", pca.explained_variance_ratio_)
print("the number of feature selected through PCA is :", pca.n_components_)
```

# data splitting

```python
# data splitting in train and test
from sklearn.model_selection import train_test_split
x_train_pca,x_test,y_train_pca,y_test=train_test_split(x_pca,y,test_size=0.25,random_state=10)
```

```python
# print the shape of splitting data
print(x_train_pca.shape,x_test.shape,y_train_pca.shape,y_test.shape)
```

#implementing model

```python
from sklearn.metrics import r2_score
```

```python
# random forest regression
from sklearn.ensemble import RandomForestRegressor

df_model = RandomForestRegressor()
df_model.fit(x_train_pca, y_train_pca)

train_pred_df = df_model.predict(x_train_pca)
train_r2 = r2_score(y_train_pca, train_pred_df)
test_pred_df = df_model.predict(x_test)
test_r2 = r2_score(y_test, test_pred_df)
# accuracy
print("Accuracy score of train data :"+str(round(100*train_r2, 4))+" %")
print("Accuracy score of test data :"+str(round(100*test_r2, 4))+" %")
```
